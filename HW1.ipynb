## 函數介紹：Leaky ReLU 函數

Leaky ReLU (Leaky Rectified Linear Unit) 是一種改進自傳統 ReLU 激活函數的變體。對於傳統的 ReLU 函數，當輸入小於零時，輸出為零；當輸入大於零時，輸出等於輸入。而 Leaky ReLU 則對負值輸入引入一個小的斜率，使函數定義為：

$$f(x) = \begin{cases}
x, & \text{if}\ x > 0 \\
\alpha x, & \text{if}\ x \leq 0
\end{cases}$$

其中 $\alpha$ 是一個很小的常數，通常設為 0.01。  

### Leaky ReLU 相較於 ReLU 的優點：

1. **解決了 "Dying ReLU" 問題**：傳統的 ReLU 在負值區域輸出為零，導致梯度也為零，可能使神經元在訓練過程中永久停止學習（即"死亡")。Leaky ReLU 通過在負值區域引入小的梯度使神經元能繼續學習。

2. **更好的梯度流動**：即使在負值區域，Leaky ReLU 仍然允許小量梯度流過，使網絡更容易收斂。

3. **保持非線性特性**：同時保持了 ReLU 計算簡便、收斂快速的優點，又增加了負數區間的表達能力。

4. **加速學習過程**：通常比傳統 ReLU 能更快地收斂，因為更少的神經元會被停止更新。

在深度學習中，Leaky ReLU 是一種常用的激活函數選擇，特別是在需要避免神經元死亡的情況下。

```python
def leaky_relu(x, alpha=0.05):
    return np.where(x > 0, x, alpha * x)

x = np.linspace(-10, 10, 400)
y = leaky_relu(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'Leaky ReLU ($\alpha=0.05$)', color='b')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Leaky ReLU Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()

plt.show()
```
